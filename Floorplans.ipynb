```python
# dependencies
import pandas as pd
import numpy as np

import datetime as dt
import matplotlib.pyplot as plt
import statistics
```

```python
# function to read csv into dataframe
def load_file(data):
    file_df = pd.read_csv(data, header = None)
    return file_df
```

```python
# test, ohd_mag_20
ohd_mag_20 = "source/OHD 2020 - 2020.csv"
ohd_mag_20_df = load_file(ohd_mag_20)
ohd_mag_20_df.head()
```

```python
# clean ohd_mag_20

# delete first row
clean_ohd_mag_20 = ohd_mag_20_df.drop([0,1])
# delete unnecessary columns
clean_ohd_mag_20 = clean_ohd_mag_20.drop(columns = [2,3,7,8,9,10,11])
# replace column headers
clean_ohd_mag_20.columns = clean_ohd_mag_20.iloc[0]
# reset index of rows
clean_ohd_mag_20 = clean_ohd_mag_20[1:].reset_index(drop = True)
# replace empty cells with NaNs
clean_ohd_mag_20.replace(r'^\s+$', np.nan, regex=True)
# fill NaN cells with ones above
clean_ohd_mag_20[['Date']] = clean_ohd_mag_20[['Date']].fillna(method = 'ffill')
clean_ohd_mag_20.loc[79:85,:]
```

```python
# further cleaning

# convert objects to correctt dtypes
clean_ohd_mag_20['Location'] = clean_ohd_mag_20['Location'].astype('string')
clean_ohd_mag_20['Time Spent (hr)'] = clean_ohd_mag_20['Time Spent (hr)'].astype('float')

# change NaN sqft to 0 then convert to integers
clean_ohd_mag_20['Square Footage'] = clean_ohd_mag_20['Square Footage'].fillna(0)
clean_ohd_mag_20['Square Footage'] = clean_ohd_mag_20['Square Footage'].astype('str')
clean_ohd_mag_20['Square Footage'] = clean_ohd_mag_20['Square Footage'].apply(lambda x : x.replace(',',''))
clean_ohd_mag_20['Square Footage'] = clean_ohd_mag_20['Square Footage'].astype('int')

# delete the '$' sign then convert to float
clean_ohd_mag_20['Time Cost + Parking'] = clean_ohd_mag_20['Time Cost + Parking'].str[1:].astype('float')
#clean_ohd_mag_20['Date'].dt.strftime('%d-%m-%y')
clean_ohd_mag_20.dtypes
```
